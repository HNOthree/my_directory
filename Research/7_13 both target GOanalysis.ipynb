```python
import pandas as pd
```

```python
miRNA1_target=pd.read_csv("/Users/miyakesho/Google_Drive/master/research/csv/miRNA1_both_target_list.csv")
miRNA155_target=pd.read_csv("/Users/miyakesho/Google_Drive/master/research/csv/miRNA155_both_target_list.csv")
refseq_to_gene_symbol=pd.read_csv("/Users/miyakesho/Google_Drive/master/research/csv/Ref_seq_to_Gene_symbol.csv")
display(miRNA1_target.head(10))
display(refseq_to_gene_symbol.head(10))
```

```python
miRNA1_ref=pd.read_csv("/Users/miyakesho/Google_Drive/master/research/csv/miRNA1_target_list.csv")
miRNA155_ref=pd.read_csv("/Users/miyakesho/Google_Drive/master/research/csv/miRNA155_target_list.csv")
miRNA1_to_refseq=pd.merge(miRNA1_ref,refseq_to_gene_symbol,on="Ref_seq")
miRNA155_to_refseq=pd.merge(miRNA155_ref,refseq_to_gene_symbol,on="Ref_seq")
display(miRNA1_to_refseq)
display(miRNA155_to_refseq)
```

```python
miRNA1_refseq_target=pd.merge(miRNA1_to_refseq,miRNA1_target,on="Gene_symbol")
miRNA155_refseq_target=pd.merge(miRNA155_to_refseq,miRNA155_target,on="Gene_symbol")
miRNA155_refseq_target
```

```python
miRNA1_refseq=miRNA1_refseq_target["Ref_seq"]
miRNA155_refseq=miRNA155_refseq_target["Ref_seq"]
miRNA1_refseq.to_csv("miRNA1_both_target_refseq.csv",index=False)
miRNA155_refseq.to_csv("miRNA155_both_target_refseq.csv",index=False)
```
