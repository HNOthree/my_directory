修飾部位に関して特異的なデータの選出

```python
import pandas as pd
```

```python
#m6AIDのデータの呼び出し
m6A_data=pd.read_csv("/Users/miyakesho/Box Sync/Personal/m6A/data/csv/phase1/all_mod_list_cell.csv")
m6A_data
```

```python
#修飾部位に関して細胞株の数を算出
cell_count=m6A_data.groupby(["mod_ID"]).count()
cell_count.head(10)
```

```python
#修飾部位に関して一つの細胞株のみ選出
unique_cell=cell_count[cell_count.chr==1]
unique_cell
```

```python
#特異的な修飾部位に関してのIDの抽出
unique_cell=unique_cell.reset_index()
unique_cell=unique_cell[["mod_ID"]]
unique_cell
```

```python
#修飾部位に関して細胞株が一種類のみのデータの抽出
new=pd.merge(unique_cell, m6A_data, on='mod_ID')
new
```

```python
#細胞株ごとの特異的な修飾部位の箇所
new.groupby("mod_cell_line").count()
```

```python
#特異的なデータの保存
new.to_csv("specific_mod_data.csv",index=False)
```

```python
#特異的なデータの修飾位開始位置をソートしたデータの保存
new=new.sort_values(by=["mod_start"], ascending=True)
new.to_csv("sorted_specific_mod_data.csv",index=False)
```
