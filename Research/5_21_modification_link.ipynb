それぞれの修飾を一つのデータにまとめる

```python
import pandas as pd
```

```python
m1A=pd.read_csv("/Users/miyakesho/Google_Drive/master/research/downloads/hg19_m1A.csv")
m5C=pd.read_csv("/Users/miyakesho/Google_Drive/master/research/downloads/hg19_m5C.csv")
m6A=pd.read_csv("/Users/miyakesho/Google_Drive/master/research/downloads/hg19_m6A.csv")
NM=pd.read_csv("/Users/miyakesho/Google_Drive/master/research/downloads/hg19_NM.csv")
pseudo=pd.read_csv("/Users/miyakesho/Google_Drive/master/research/downloads/hg19_PseudoU.csv")
m1A.head(10)
```

```python
m5C.head(10)
```

```python
modification=pd.concat([m1A,m5C,m6A,NM,pseudo])
modification.head(10)
```

```python
modification=modification[["mod_id","chromosome","gene_name","gene_type","mod_start","mod_end","mod_type","motif_sequence","pubmed_ids","region","strand","support_list","support_number"]]
modification.head(10)
```

```python
modification.to_csv("modification_list.csv",index=False)
```
