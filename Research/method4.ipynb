GO解析の結果を解析

```python
import pandas as pd
```

```python
cell=["A549","HeLa","Huh7","HepG2","MONO-MAC-6","type Jurkat","U2OS"]
```

```python
#GO解析の結果を呼び出し,全結合
GO_term=[]
for cell_line in cell:
    cell_line_list=pd.read_csv("/Users/miyakesho/Box Sync/Personal/m6A/data/csv/phase1/GO/{}GO.csv".format(cell_line))
    cell_line_list=cell_line_list[["GO_term","Pvalue"]]
    #cell_line_list=pd.DataFrame(cell_line_list.values,columns=["GO_term","Pvalue"])
    for i in range(0,len(cell_line_list)):
        a_col=[]
        col=cell_line_list.iloc[i]
        term=col["GO_term"]
        pvalue=col["Pvalue"]
        a_col.append(term)
        a_col.append(pvalue)
        GO_term.append(a_col)
    
```

```python
#データの成形
GO_term=pd.DataFrame(GO_term,columns=["GO_term","Pvalue"])
GO_term
```

```python
#全結合したGO termの出現回数をカウント
term_list=GO_term.groupby(["GO_term"]).count()
term_list.to_csv("GO_term_cellline.csv")
```

```python
#全ての細胞株で見られたGO termのリスト
common_term=term_list[term_list.Pvalue==7]
common_term=common_term.reset_index()
common_term=pd.DataFrame(common_term.values,columns=["GO_term","cell_line_kind"])
common_term
```

```python
common_term.to_csv("common_GO_term.csv",index=False)
```

```python
#一つの細胞株でしか見られなかったGO termのリスト
unique_term=term_list[term_list.Pvalue==1]
unique_term=unique_term.reset_index()
unique_term=pd.DataFrame(unique_term.values,columns=["GO_term","cell_line_kind"])
unique_term
```

```python
#細胞株ごとにユニークなGOtermの保存
for cell_line in cell:
    cell_line_list=pd.read_csv("/Users/miyakesho/Box Sync/Personal/m6A/data/csv/phase1/GO/{}GO.csv".format(cell_line))
    #cell_line_list=cell_line_list[["GO_term","Pvalue"]]
    uni=pd.merge(cell_line_list,unique_term,on="GO_term")
    uni.to_csv("{}_unique_GOlist.csv".format(cell_line),index=False)
```

```python
#GOのIDの削除
import pandas as pd
cell=["A549","HeLa","Huh7","HepG2","MONO-MAC-6","type Jurkat","U2OS"]
```

```python
#細胞株ごとに削除
import re
for cell_line in cell:
    ID_list=pd.read_csv("/Users/miyakesho/Box Sync/Personal/m6A/data/csv/phase1/GO/unique/{}_unique_GOlist.csv".format(cell_line))
    ID=[]
    for i in range(0,len(ID_list)):
        a_col=[]
        col=ID_list.iloc[i]
        GO=col.GO_term
        GO=re.sub("GO:\d+","",GO)
        GO=GO.replace("()","")
        pvalue=col.Pvalue
        a_col.append(GO)
        a_col.append(pvalue)
        ID.append(a_col)
    ID=pd.DataFrame(ID,columns=["GO_term","Pvalue"])
    ID.to_csv("{}GO_unique_excludeID.csv".format(cell_line),index=False)
```
