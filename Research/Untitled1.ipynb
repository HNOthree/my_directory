```python
reader=pd.read_csv("/Users/miyakesho/Box Sync/Personal/m6A/data/phase2/Reader/statistical_decrease_Reader.csv")
writer=pd.read_csv("/Users/miyakesho/Box Sync/Personal/m6A/data/phase2/Writer/GSE63591_D-METTL3-Ribosome_profiling.csv")
```

```python
reader_target_gene=[]
for gene in reader["Gene symbol"]:
    reader_target_gene.append(gene)
reader_target_gene
```

```python
reader_target=[]
non_target=[]
for i in range(0,len(writer)):
    col=writer.iloc[i]
    gene=col["Gene symbol"]
    if gene not in reader_target_gene:
        non_target.append(col)
    elif gene in reader_target_gene:
        reader_target.append(col)
    else:
        print("Error")

reader_target=pd.DataFrame(reader_target)
non_target=pd.DataFrame(non_target)
non_target
```

```python
import numpy as  np
from matplotlib import pyplot as plt
import seaborn as sns

sns.set()
sns.set_style('whitegrid')
sns.set_palette('gray')

np.random.seed(2018)

x1 = np.random.normal(10, 2, 20)
x2 = np.random.normal(15, 3, 20)
x3 = np.random.normal(5, 1, 20)

fig = plt.figure()
ax = fig.add_subplot(1, 1,1)
ax.boxplot([non_target["TE_log2(siMETTL3/siControl)"], reader_target["TE_log2(siMETTL3/siControl)"]], labels=['Control', 'Reader_target'])
#ax.set_xlabel('tissue')
ax.set_ylabel('Translation Efficiency')
ax.set_ylim(-5, 5)

plt.show()
```
