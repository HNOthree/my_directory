miRNAのターゲットとされている発現効率を比較

```python
import pandas as pd
```

```python
ID_list=pd.read_csv("/Users/miyakesho/Google ドライブ/Personal/m6A/data/phase2/Eraser/Eraser_logfold.csv")
miRNA1_list=pd.read_csv("/Users/miyakesho/Google ドライブ/Personal/m6A/data/phase2/miRNA/mir1_logfold.csv")
miRNA155_list=pd.read_csv("/Users/miyakesho/Google ドライブ/Personal/m6A/data/phase2/miRNA/mir155_logfold.csv")
```

```python
ID_list=ID_list[["RefseqAccession","Symbol"]]
ID_list.head(10)
```

```python
miRNA1_list.head(10)
```

```python
miRNA1_list=pd.merge(ID_list,miRNA1_list,on="RefseqAccession")
miRNA1_list = miRNA1_list.rename(columns={'Symbol': 'Gene symbol'})
miRNA155_list=pd.merge(ID_list,miRNA155_list,on="RefseqAccession")
miRNA155_list = miRNA155_list.rename(columns={'Symbol': 'Gene symbol'})
miRNA1_list.head(10)
miRNA_list=pd.concat([miRNA1_list, miRNA155_list])
miRNA_list[~miRNA_list.duplicated()]
miRNA_list = miRNA_list.rename(columns={'Symbol': 'Gene symbol'})
#miRNA.append()
miRNA_list
```

```python
reader_target=pd.read_csv("/Users/miyakesho/Google ドライブ/Personal/m6A/data/phase2/Reader/statistical_decrease_Reader.csv")
reader_target.head(10)
```

```python
merge=pd.merge(miRNA155_list,reader_target,on="Gene symbol")
merge
```

```python
both=pd.merge(miRNA1_list[miRNA1_list["mi_target"]=="mir1_target"],miRNA155_list[miRNA155_list["mi_target"]=="mir155_target"],on="RefseqAccession")
both_merge=pd.merge(both,ID_list,on="RefseqAccession")
both_merge = both_merge.rename(columns={'Symbol': 'Gene symbol'})
both_merge=pd.merge(both_merge,reader_target,on="Gene symbol")
both_merge
```

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
sns.boxplot(x='mi_target', y='TE_log2(siYTHDF1/siControl)ave', data=merge, showfliers=False, ax=ax)
sns.stripplot(x='mi_target', y='TE_log2(siYTHDF1/siControl)ave', data=merge, jitter=True, color='black', ax=ax)
plt.savefig("miRNA155_target.png",dpi=300)
plt.show()
```
