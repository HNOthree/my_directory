```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
num = random.uniform(2.0,5.0)
both_x = np.random.rand(100)*4-3
both_y = np.random.rand(100)*4-3
YTHDF_x=np.random.rand(100)*3-3
YTHDF_y=np.random.rand(100)*3-1.5
miRNA_x=np.random.rand(100)*3-1.5
miRNA_y=np.random.rand(100)*3-3
control_x=np.random.rand(150)*6-3
control_y=np.random.rand(150)*6-3
```

```python
fig = plt.figure()

ax = fig.add_subplot(1,1,1)

ax.scatter(both_x,both_y, c='red',label="both target")
ax.scatter(YTHDF_x,YTHDF_y, c='green',label="YTHDF1_target")
ax.scatter(miRNA_x,miRNA_y, c='blue',label="miRNA_target")
ax.scatter(control_x,control_y, c='yellow',label="No target")
ax.set_xlim(-3.5,3.5)
ax.set_ylim(-3.5, 3.5)
ax.set_title('first scatter plot')
ax.set_xlabel('log(siYTHDF1/control)')
ax.set_ylabel('log(over expression miRNA/control)')
ax.legend(loc='upper right')
fig.show()
```
