#両方の実験データから共通で出てきた遺伝子を選出

```python
m6A_data=pd.read_csv("/Users/miyakesho/Google_Drive/master/research/downloads/GSE63591_C-Y1-Ribosome_profiling.csv")
miRNA_1_data=pd.read_csv("/Users/miyakesho/Google_Drive/master/research/csv/symboled_miRNA1.csv")
miRNA_155_data=pd.read_csv("/Users/miyakesho/Google_Drive/master/research/csv/symboled_miRNA155.csv")
display(m6A_data.head(10))
display(miRNA_1_data.head(10))
display(miRNA_155_data.head(10))
```

```python
m6A_miRNA1=pd.merge(m6A_data,miRNA_1_data,on="Gene_symbol")
m6A_miRNA155=pd.merge(m6A_data,miRNA_155_data,on="Gene_symbol")
display(m6A_miRNA1.head(10))
print("m6A_miRNA1 gene:",len(m6A_miRNA1))
display(m6A_miRNA155.head(10))
print("m6A_miRNA155 gene",len(m6A_miRNA155))
```
