```python
import pandas as pd
```

```python
m6A_list=pd.read_csv("/Users/miyakesho/Google_Drive/bachelor/m6A/data/rawdata/hg19_m6A.csv")
m6A_list.head(10)
```

```python
m6A_list.groupby('gene_name').count().head(10)
```

```python
count=0
gene_list=[]
for pre_gene in m6A_list.gene_name:
    if (count%100000==0):
        print(count,"/",len(m6A_list))
    genes= pre_gene.split(",")
    for gene in genes:
        if (gene not in gene_list):
            gene_list.append(gene)
    count+=1
```
