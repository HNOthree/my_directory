```python
def get_seq(chromosome,start,end,strand,window):
    test_data = open("/Users/miyakesho/Google_Drive/master/research/txt/{}.txt".format(chromosome), "r")
    sequence=test_data.read()
    target_seq=sequence[start-window:end+window]
    target_seq=bigletter(target_seq)
    if(strand=="-"):
        target_seq=convert(target_seq)
    return target_seq
```

```python
def bigletter(seq):
    new_seq=""
    for i in seq:
        if(i=="a"):
            new_seq+="A"
        elif(i=="g"):
            new_seq+="G"
        elif(i=="c"):
            new_seq+="C"
        elif(i=="t"):
            new_seq+="T"
        else:
            new_seq+=i
    return new_seq
```

```python
def convert(sequence):
    
    i=len(sequence)-1
    convert_seq=""
    while(i>=0):
        letter=sequence[i]
        if(letter=="A"):
            convert_seq+="T"
        elif(letter=="T"):
            convert_seq+="A"
        elif(letter=="G"):
            convert_seq+="C"
        elif(letter=="C"):
            convert_seq+="G"
        else:
            print("ERROR")
        i-=1
    
   
    return convert_seq
```

```python
def for_miRNA(seq):
    new_seq=""
    for i in seq:
        if(i=="A"):
            new_seq+="U"
        elif(i=="T"):
            new_seq+="A"
        elif(i=="G"):
            new_seq+="C"
        elif(i=="C"):
            new_seq+="G"
    new_seq=new_seq[::-1]
    return new_seq
```

```python
def miRNA_combine(sequence,miRNA):
    seed=miRNA[1:8]
    match=[]
    location=[]
    for i in range(0,len(sequence)-len(seed)+1):
        candidate=for_miRNA(sequence[i:i+7])
        
        if(seed==candidate):
            print(sequence,candidate,seed)
            location.append(round(i/len(sequence)))
    
    for j in location:
        if(j==1):
            match.append("downstream")
        else:
            match.append("upstream")
    return match
```
