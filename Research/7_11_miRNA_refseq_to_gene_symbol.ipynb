miRNAのデータのRefseqの番号をgenesymbolに書き換え

```python
import pandas as pd
```

```python
miRNA_1=pd.read_csv("/Users/miyakesho/Google_Drive/master/research/csv/miRNA1.csv")
display(miRNA_1.head(10))
print("gene",len(miRNA_1))
```

```python
ref_symbol=pd.read_csv("/Users/miyakesho/Google_Drive/master/research/csv/Ref_seq_to_Gene_symbol.csv")
ref_symbol.head(10)
```

```python
merge_list=pd.merge(miRNA_1, ref_symbol, on='Ref_seq', how='left')
merge_list=merge_list[~merge_list.duplicated()]
merge_list.to_csv("miRNA_1")
print(merge_list)
```
