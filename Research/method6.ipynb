miRNAのターゲットとオーバーラップしているm6Aの修飾を持つ遺伝子の抽出

```python
import pandas as pd
```

```python
miRNA_mod_list=pd.read_csv("/Users/miyakesho/Box Sync/Personal/m6A/data/csv/phase1/specific_mod_mitar.csv")
miRNA_mod_list
```

```python
miRNA_mod_list.groupby(["cell_line"]).count()
```

```python
#cell lineの抽出
cell_list=[]
for term in miRNA_mod_list.cell_line:
    if term not in cell_list:
        cell_list.append(term)
        
        
cell_list
```

```python
for cell in cell_list:
    specific_mod=miRNA_mod_list[miRNA_mod_list.cell_line==cell]
    gene_list=[]
    for gene in specific_mod.Gene_ID:
        if gene not in gene_list:
            gene_list.append(gene)
    gene_list=pd.DataFrame(gene_list)
    gene_list.to_csv("{}_miTar_mod_gene.csv".format(cell),index=False,header=False)
```

```python
#GO解析の結果を呼び出し,全結合
GO_term=[]
for cell_line in cell_list[:4]:
    cell_line_list=pd.read_csv("/Users/miyakesho/Box Sync/Personal/m6A/data/csv/phase1/GO/miRNA/{}_miRNA_GO.csv".format(cell_line))
    cell_line_list=cell_line_list[["GOterm","Pvalue"]]
    #cell_line_list=pd.DataFrame(cell_line_list.values,columns=["GO_term","Pvalue"])
    for i in range(0,len(cell_line_list)):
        a_col=[]
        col=cell_line_list.iloc[i]
        term=col["GOterm"]
        pvalue=col["Pvalue"]
        a_col.append(term)
        a_col.append(pvalue)
        GO_term.append(a_col)
    
```

```python
GO_term=pd.DataFrame(GO_term,columns=["GOterm","Pvalue"])
```

```python
GO_list=GO_term.groupby(["GOterm"]).count()
```

```python
common=GO_list[GO_list.Pvalue>1]
common.to_csv("common_miRNA_GO.csv")
```

```python
unique=GO_list[GO_list.Pvalue==1]
unique=unique.reset_index()
unique
```

```python
#GO解析の結果を呼び出し,全結合
GO_term=[]
for cell_line in cell_list[:4]:
    cell_line_list=pd.read_csv("/Users/miyakesho/Box Sync/Personal/m6A/data/csv/phase1/GO/miRNA/{}_miRNA_GO.csv".format(cell_line))
    cell_line_list=cell_line_list[["GOterm","Pvalue"]]
    
    specific_unique_GO=pd.merge(cell_line_list,unique,on="GOterm")
    specific_unique_GO.to_csv("{}_specific_miRNA_GO.csv".format(cell_line),index=False)
```

```python
#IDリストの削除
#細胞株ごと
import re
import pandas as pd
cell=["A549","HeLa","Huh7","HepG2"]
for cell_line in cell:
    ID_list=pd.read_csv("/Users/miyakesho/Box Sync/Personal/m6A/data/csv/phase1/GO/miRNA/{}_specific_miRNA_GO.csv".format(cell_line))
    ID=[]
    for i in range(0,len(ID_list)):
        a_col=[]
        col=ID_list.iloc[i]
        GO=col.GOterm
        GO=re.sub("GO:\d+","",GO)
        GO=GO.replace("()","")
        pvalue=col.Pvalue_x
        a_col.append(GO)
        a_col.append(pvalue)
        ID.append(a_col)
    ID=pd.DataFrame(ID,columns=["GO_term","Pvalue"])
    ID.to_csv("miRNA_{}GO_unique_excludeID.csv".format(cell_line),index=False)
```
