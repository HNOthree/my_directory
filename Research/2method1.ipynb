miRNAとm6Aの競合部位を選出

```python
import pandas as pd
```

```python
miRNA_target=pd.read_csv("/Users/miyakesho/Box Sync/Personal/m6A/data/rawdata/hg19_all_mod_mirTar.csv")
miRNA_target
```

```python
m6A_target=miRNA_target[miRNA_target.modType=="m6A"]
m6A_target
```

```python
m6A_target.to_csv("m6A_miRNA_target.csv",index=False)
```

```python
m6A_list=m6A_target[["mirTar_Id","modId","geneName","modStart","modEnd","mirTarStart","mirTarEnd","mirTarName"]]
#m6A_list=pd.DataFrame(m6A_list.values,columns=["mirTar_Id","modId","Gene symbol","modStart","modEnd","mirTarStart","mirTarEnd","mirTarName"])
m6A_list
```

```python
header_list=[]

for term in m6A_list:
    header_list.append(term)
    
header_list.append("Gene Symbol")
splitlist=[]
for i in range(0,len(m6A_list)):
    col=m6A_list.iloc[i]
    symbol=col["geneName"].split(",")
    for gene in symbol:
        a_col=[]
        for term in col:
            a_col.append(term)
        a_col.append(gene)
        splitlist.append(a_col)
```

```python
splitlist=pd.DataFrame(splitlist)
splitlist.columns=header_list
splitlist
```

```python
splitlist=splitlist[~splitlist.duplicated()]
splitlist=splitlist.drop("geneName", axis=1)
splitlist
```

```python
splitlist.to_csv("miRNA_target.csv",index=False)
```
