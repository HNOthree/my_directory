```python
import pandas as pd
from scipy import stats
```

```python
#YTHDF の翻訳効率 miRNA1、mIRNA155の翻訳効率のデータ取得
m6A=pd.read_csv("/Users/miyakesho/Google_Drive/master/research/downloads/GSE63591_C-Y1-Ribosome_profiling.csv")
miRNA_1=pd.read_csv("/Users/miyakesho/Google_Drive/master/research/csv/symboled_miRNA1.csv")
miRNA_155=pd.read_csv("/Users/miyakesho/Google_Drive/master/research/csv/symboled_miRNA155.csv")
display(m6A.head(10))
display(miRNA_1.head(10))
```

```python
#m6AとmiRNAのデータをそれぞれ結合
m6A_miRNA1_merge_list=pd.merge(m6A, miRNA_1, on='Gene_symbol')
m6A_miRNA155_merge_list=pd.merge(m6A, miRNA_155, on='Gene_symbol')
print(len(m6A_miRNA1_merge_list),len(m6A_miRNA155_merge_list))
```

```python
display(m6A_miRNA1_merge_list.head(10))
display(m6A_miRNA155_merge_list.head(10))
```

```python
miRNA1_target=[]
miRNA155_target=[]
YTHDF_target=[]
miRNA1_target_df=pd.read_csv("/Users/miyakesho/Google_Drive/master/research/csv/miRNA1_target_list.csv")
miRNA155_target_df=pd.read_csv("/Users/miyakesho/Google_Drive/master/research/csv/miRNA155_target_list.csv")
YTHDF_target_df=pd.read_csv("/Users/miyakesho/Google_Drive/master/research/csv/PARCLIP_target.csv")
for i in miRNA1_target_df.Ref_seq:
    miRNA1_target.append(i)
for i in miRNA155_target_df.Ref_seq:
    miRNA155_target.append(i)

for i in YTHDF_target_df["Gene symbol"]:
    YTHDF_target.append(i)
YTHDF_target[0:10]
```

```python
 ######################miRNA1 ################
```

```python
#miRNA1 vs m6A　

miRNA1_both_target_pd=[]
miRNA1_target_pd=[]
miRNA1_YTHDF_target_pd=[]
miRNA1_non_target_pd=[]
for i in range(0,len(m6A_miRNA1_merge_list)):
    gene=m6A_miRNA1_merge_list.iloc[i]
    ref_seq=gene.Ref_seq
    symbol=gene.Gene_symbol
    m6A_TE=gene["TE_log2(siYTHDF1/siControl)ave"]
    miRNA_TE=gene["miRNA_TranslEff_log2fc"]
    m6A_RNAseq=gene["polyA_log2(siYTHDF1/siControl)ave"]
    miRNA_RNAseq=gene["miRNA_rnaseq_log2fc"]
    m6A_ribo=gene["ribo_log2(siYTHDF1/siControl)ave"]
    miRNA_ribo=gene["miRNA_footprint_log2fc"]
    
    row=[m6A_TE,miRNA_TE,m6A_RNAseq,miRNA_RNAseq,m6A_ribo,miRNA_ribo]
    if (ref_seq in miRNA1_target and symbol in YTHDF_target ):
        miRNA1_both_target_pd.append(row)
    elif(ref_seq in miRNA1_target):
        miRNA1_target_pd.append(row)
    elif(symbol in YTHDF_target):
        miRNA1_YTHDF_target_pd.append(row)
    else:
        miRNA1_non_target_pd.append(row)
```

```python
miRNA1_both_target_pd=pd.DataFrame(miRNA1_both_target_pd,columns=["YTHDF1_TE","miRNA1_TE","YTHDF1_RNAseq","miRNA1_RNAseq","YTHDF1_ribo","miRNA1_ribo"])
miRNA1_target_pd=pd.DataFrame(miRNA1_target_pd,columns=["YTHDF1_TE","miRNA1_TE","YTHDF1_RNAseq","miRNA1_RNAseq","YTHDF1_ribo","miRNA1_ribo"])
miRNA1_YTHDF_target_pd=pd.DataFrame(miRNA1_YTHDF_target_pd,columns=["YTHDF1_TE","miRNA1_TE","YTHDF1_RNAseq","miRNA1_RNAseq","YTHDF1_ribo","miRNA1_ribo"])
miRNA1_non_target_pd=pd.DataFrame(miRNA1_non_target_pd,columns=["YTHDF1_TE","miRNA1_TE","YTHDF1_RNAseq","miRNA1_RNAseq","YTHDF1_ribo","miRNA1_ribo"])
```

```python
miRNA1_TE_YTHDF1_result = stats.mannwhitneyu(miRNA1_both_target_pd.YTHDF1_TE,miRNA1_non_target_pd.YTHDF1_TE)
miRNA1_TE_miRNA1_result = stats.mannwhitneyu(miRNA1_both_target_pd.miRNA1_TE,miRNA1_non_target_pd.miRNA1_TE)
miRNA1_TE_YTHDF1_result = stats.mannwhitneyu(miRNA1_both_target_pd.YTHDF1_TE,miRNA1_non_target_pd.YTHDF1_TE)
miRNA1_TE_miRNA1_result = stats.mannwhitneyu(miRNA1_both_target_pd.miRNA1_TE,miRNA1_non_target_pd.miRNA1_TE)
```

```python
def test1(miRNA1_both_target_pd,miRNA1_non_target_pd):
    topics=["YTHDF1_TE","miRNA1_TE","YTHDF1_RNAseq","miRNA1_RNAseq","YTHDF1_ribo","miRNA1_ribo"]
    for i in topics:
        result=stats.mannwhitneyu(miRNA1_both_target_pd[i],miRNA1_non_target_pd[i])
        result=result.pvalue
        miRNA_result.append([i,result])
        print(i,result)
    return miRNA_result
```

```python
print("both_target")
both_target_test=test1(miRNA1_both_target_pd,miRNA1_non_target_pd)
print("\nYTHDF1_target")
YTHDF_target_test=test1(miRNA1_YTHDF_target_pd,miRNA1_non_target_pd)
print("\nmiRNA1_target")
miRNA1_target_test=test1(miRNA1_target_pd,miRNA1_non_target_pd)
print("\n both vs miRNA target")
both_vs_miRNA1_target_test=test1(miRNA1_both_target_pd,miRNA1_target_pd)
print("\n both vs YTHDF target")
both_vs_miRNA1_target_test=test1(miRNA1_both_target_pd,miRNA1_YTHDF_target_pd)
```

```python
def test155(miRNA155_both_target_pd,miRNA155_non_target_pd):
    topics=["YTHDF1_TE","miRNA155_TE","YTHDF1_RNAseq","miRNA155_RNAseq","YTHDF1_ribo","miRNA155_ribo"]
    for i in topics:
        result=stats.mannwhitneyu(miRNA155_both_target_pd[i],miRNA155_non_target_pd[i])
        result=result.pvalue
        miRNA_result.append([i,result])
        print(i,result)
    return miRNA_result
```
