hg19.fa.outのデータ整形

```python
import pandas as pd 
```

```python
input_file_name="/Users/miyakesho/Google_Drive/master/research/downloads/hg19.fa.out"
matrix=[]
with open(input_file_name,"r") as f:
    for i in f.readlines():
        i=i.split(" ")
        col=[]
        for j in i:
            if (j!=""):
                col.append(j)
        matrix.append(col)
```

```python
matrix
```

```python
matrix=pd.DataFrame(matrix)
```

```python
matrix.head(10)
```

```python
output=matrix.iloc[3:,[4,5,6,8,9,10]]
output=pd.DataFrame(output.values,columns=["Chromosome","Start","End","Strand","Repeat","Class_Family"])
```
