```python
import pandas as pd
import numpy as np
```

```python
PARCLIP1=pd.read_csv("/Users/miyakesho/Google_Drive/master/research/csv/raw_PARCLIP1.csv")
sample=PARCLIP1.head(10)
sample
```

```python
new_PARCLIP=[]
for i in range(len(PARCLIP1)):
    row=PARCLIP1.iloc[i]
    gene_info=row[0:5].values
    for j in range(5,len(row),6):
        if(row[j]==""):
            break
        else:
            info=[]
            for k in gene_info:
                info.append(k)
            for k in row[j:j+5].values:
                info.append(k)
            new_PARCLIP.append(info)
```

```python
new_PARCLIP=pd.DataFrame(new_PARCLIP)
new_PARCLIP=new_PARCLIP.dropna(how='any')
new_PARCLIP=pd.DataFrame(new_PARCLIP.values,columns=["Gene_symbol","Name","Peak_sum","Mutation_sum","Chromosome","Start","End","Strand","Peak_mutation","Region"])
new_PARCLIP.to_csv("PARCLIP_2.csv",index=False)
```
