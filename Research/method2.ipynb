遺伝子ごとに特異的な修飾箇所のカウント

```python
import pandas as pd
```

```python
#特異的なm6Aデータの呼び出し
m6A_data=pd.read_csv("/Users/miyakesho/Box Sync/Personal/m6A/data/csv/phase1/specific_mod_data.csv")
m6A_data=m6A_data[["mod_ID","mod_cell_line"]]
m6A_data.to_csv("site_cell.csv",index=False)
```

```python
#広く発現している遺伝子に座標がオーバーラップしているデータの呼び出し
mod_to_gene=pd.read_csv("/Users/miyakesho/Box Sync/Personal/m6A/data/csv/phase1/phase1_mod_list_to_gene.csv")
mod_to_gene
```

```python
#修飾部位が特異的なデータの中から広く発現している遺伝子の中で修飾が特異的なデータの選出
specific=pd.merge(m6A_data, mod_to_gene, on='mod_ID')
specific
```

```python
specific.groupby(["mod_cell_line"]).count()
```

```python
#遺伝子のリスト選出
gene_list=[]
for gene in specific.Gene_ID:
    if gene not in gene_list:
        gene_list.append(gene)
gene_list=pd.DataFrame(gene_list,columns=["Gene_ID"])
gene_list
```

```python
import numpy as np
```

```python
#修飾カウントリスト作成
gene_count_list=np.zeros((len(gene_list),7))
gene_count_list=pd.DataFrame(gene_count_list,columns=["HeLa","HepG2","A549","U2OS","Huh7","MONO-MAC-6","type Jurkat"])
gene_count_list=pd.concat([gene_list,gene_count_list],axis=1)
gene_count_list=gene_count_list.set_index(["Gene_ID"])
gene_count_list
```

```python
#遺伝子ごとに組織特異的な修飾箇所をカウントする

for gene in gene_list.Gene_ID:
    candidates=specific[specific.Gene_ID==gene]
    
    #遺伝子における細胞株別の修飾箇所のカウント
    count_list=candidates.groupby(["mod_cell_line"]).count()
    count_list=count_list.reset_index()
    count_list=count_list[["mod_cell_line","mod_ID"]]
    
    #細胞株ごとに修飾箇所を更新
    for i in range(0,len(count_list)):
        a_loc=count_list.iloc[i]
        cell=a_loc.mod_cell_line
        count=a_loc.mod_ID
        #gene_count_list[cell][gene]=count
        gene_count_list.at[gene, cell] = count
        #print(gene_count_list)
    #print(count_list)
```

```python
gene_count_list
```

```python
#細胞株ごとに組織特異的に修飾がなされた遺伝子群の選出
for cell in gene_count_list:
    specific_gene_list=gene_count_list[gene_count_list[cell]>0]
    specific_gene_list=specific_gene_list.reset_index()
    specific_gene_list=specific_gene_list[["Gene_ID",cell]]
    specific_gene_list.to_csv("{}_specific_gene_list.csv".format(cell),index=False)
    #print(specific_gene_list)
    
```
