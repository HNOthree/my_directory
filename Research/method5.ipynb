miRNAとオーバーラップするm6Aのデータを選出する

```python
import pandas as pd
```

```python
#miRNAのデータ選出
miRNA_list=pd.read_table("/Users/miyakesho/Box Sync/Personal/m6A/data/rawdata/RMBase_hg19_all_mod_mirTar.txt",sep="\t")
m6A_miRNA_list=miRNA_list[miRNA_list.modType=="m6A"]
m6A_miRNA_list
```

```python
#実験で用いた細胞株のデータを呼び出す
exp_list=pd.read_csv("/Users/miyakesho/Box Sync/Personal/m6A/data/csv/miRNA/supprtID_cell_line.csv")
exp_list
```

```python
#必要なデータだけ呼び出す
a=m6A_miRNA_list[["mirTarId","supportList"]]
```

```python
#ターゲット箇所と実験IDの結びつけ
each_list=[]
for i in range(0,len(a)):
    if (i%1000==0):
        print(i,len(a))
    col=a.iloc[i]
    ID=col.mirTarId
    
    support=col.supportList.split(",")
    for supportID in support:
        #print(supportID)
        a_col=[]
        a_col.append(ID)
        a_col.append(supportID)
        each_list.append(a_col)

each_list=pd.DataFrame(each_list)
each_list=pd.DataFrame(each_list.values,columns=["mirTarId","support_ID"])
each_list
```

```python
#各miRNA部位に関しての細胞株のデータを結びつけ
miRNA_cell=pd.merge(each_list, exp_list, on='support_ID')

merge=pd.merge(m6A_miRNA_list, miRNA_cell, on='mirTarId')
support_drop=merge.drop("support_ID", axis=1)
support_drop=support_drop[~support_drop.duplicated()]
support_drop
```

```python
support_drop.to_csv("mi_tar_cell.csv",index=False)
```
