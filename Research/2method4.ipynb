```python
#11412遺伝子
writer_data=pd.read_csv("/Users/miyakesho/Box Sync/Personal/m6A/data/phase2/Writer/GSE63591_D-METTL3-Ribosome_profiling.csv")
writer_data.head(10)
```

```python
#194のターゲット遺伝子
reader_target=pd.read_csv("/Users/miyakesho/Box Sync/Personal/m6A/data/phase2/Reader/statistical_decrease_Reader.csv")
reader_target.head(10)
```

```python
target_gene=[]
for i in reader_target["Gene symbol"]:
    target_gene.append(i)
target_gene
```

```python
header_col=[]
for term in writer_data:
    header_col.append(term)

target=[]
non_target=[]

for i in range(0,len(writer_data)):
    a_col=[]
    col=writer_data.iloc[i]
    gene=col["Gene symbol"]

    
    if gene in target_gene:
        
        for term in col:
            a_col.append(term)
        target.append(a_col)
    elif gene not in target_gene:
        for term in col:
            a_col.append(term)
        non_target.append(a_col)
    else :
        print("Error")
        
```

```python
target=pd.DataFrame(target)
non_target=pd.DataFrame(non_target)
target.columns=header_col
non_target.columns=header_col
```

```python
import numpy as  np
from matplotlib import pyplot as plt
import seaborn as sns

sns.set()
sns.set_style('whitegrid')
sns.set_palette('gray')

np.random.seed(2018)


fig = plt.figure()
ax = fig.add_subplot(1, 1,1)
ax.boxplot([target["TE_log2(siMETTL3/siControl)"],non_target["TE_log2(siMETTL3/siControl)"]], labels=['Reader_target','Control'])
#ax.set_xlabel('tissue')
ax.set_ylabel('Translation Efficiency')
ax.set_ylim(-5, 5)
plt.savefig("Translation_efficiency_box.png",dpi=300)
plt.show()
```

```python
import numpy as np
fig, ax1 = plt.subplots()

# ヒストグラムプロットとビン情報の取得
n3, bins3, patches3 = ax1.hist(target["TE_log2(siMETTL3/siControl)"], alpha=0.7,range=(-4,4),bins=30, color='r',normed=True)

#n1, bins1, patches1 = ax1.hist(t_middle.log_length, alpha=0.7,range=(1,4),bins=50, color='b',normed=True)
n2, bins2, patches2 = ax1.hist(non_target["TE_log2(siMETTL3/siControl)"], alpha=0.7,range=(-4,4),bins=30, color='g',normed=True)
plt.xlabel("Translation Efficiency")
plt.ylabel("Frequency")


# 第2軸用値の算出
#y2 = np.add.accumulate(n1) / n1.sum()
#x2 = np.convolve(bins1, np.ones(2) / 2, mode="same")[1:]
y3 = np.add.accumulate(n2) / n2.sum()
x3 = np.convolve(bins2, np.ones(2) / 2, mode="same")[1:]
y4 = np.add.accumulate(n3) / n3.sum()
x4 = np.convolve(bins3, np.ones(2) / 2, mode="same")[1:]

# 第2軸のプロット
ax2 = ax1.twinx()
lines3 = ax2.plot(x4, y4, ls='--', color='r', marker='o',
         label='Reader Targets')

#lines1 = ax2.plot(x2, y2, ls='--', color='b', marker='o',
  #       label='Widely Methylated')
lines2 = ax2.plot(x3, y3, ls='--', color='g', marker='o',
         label='Control')
plt.ylabel("Cumulative ratio")
ax2.grid(visible=False)
plt.legend()


plt.savefig("Translation_efficiency.png",dpi=300)
plt.show()
```
