```python
overlap=pd.read_csv("/Users/miyakesho/Google_Drive/master/research/csv/miRNA1_{}overlap_count.csv".format(50))
overlap.head(10)
```

```python
miRNA1=pd.read_csv("/Users/miyakesho/Google_Drive/master/research/csv/symboled_miRNA1.csv")
miRNA1_target=pd.read_csv("/Users/miyakesho/Google_Drive/master/research/csv/miRNA1_target_list.csv")
miRNA1_target_df=pd.concat([miRNA1,miRNA1_target], axis=1, join='inner')
miRNA1_target_df.head(10)
```

```python
miRNA1list=[]
for i in miRNA1_target_df.Gene_symbol:
    if i not in miRNA1list:
        miRNA1list.append(i)
len(miRNA1list)
```

```python
overlap_list=[]
non_list=[]
for i in overlap.Gene_symbol:
    if i  in miRNA1list:
        overlap_list.append(i)
    else:
        non_list.append(i)
```

```python
sequence_match=pd.merge(overlap, miRNA1_target_df, on='Gene_symbol', how='inner')
sequence_match
```

```python
print(len(overlap_list),len(non_list))
```

```python
import matplotlib.pyplot as plt
plt.hist(overlap.location,bins=100,cumulative=False)
plt.show()
```

```python
a=[]
gene=[]
for i in range(0,len(sequence_match)) :
    row=sequence_match.iloc[i]
    if row.Gene_symbol not in gene:
        gene.append(row.Gene_symbol)
        a.append(row.miRNA_footprint_log2fc)
plt.boxplot(a)
plt.show()
plt.boxplot(miRNA1_target_df.miRNA_footprint_log2fc)
plt.show()
```
