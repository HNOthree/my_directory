miRNAのエクセルのデータをcsvに加工

```python
import pandas as pd
```

```python
file_path="/Users/miyakesho/Box Sync/Personal/m6A/data/phase2/miRNA/GSE21992_mir155_summaryTable.txt"
```

```python
file_txt=pd.read_csv(file_path,sep="\t", skiprows=2)
file_txt
```

```python
file_text=file_txt[["RefseqAccession","noSite_translEff_log2fc",">=1site_translEff_log2fc"]]
```

```python
file_text
```

```python
miRNA=[]
no_target=[]
for i in range(0,len(file_text)):
    col=file_text.iloc[i]
    a_col=[]
    if (col["noSite_translEff_log2fc"]!="-"):
        a_col.append(col.RefseqAccession)
        a_col.append(float(col["noSite_translEff_log2fc"]))
        no_target.append(a_col)
for i in range(0,len(file_text)):
    col=file_text.iloc[i]
    a_col=[]
    if (col[">=1site_translEff_log2fc"]!="-"):
        a_col.append(col.RefseqAccession)
        a_col.append(float(col[">=1site_translEff_log2fc"]))
        miRNA.append(a_col)
```

```python
miRNA=pd.DataFrame(miRNA,columns=["RefseqAccession","logfold_TE"])
no_target=pd.DataFrame(no_target,columns=["RefseqAccession","logfold_TE"])
```

```python
#log戻す
nonlog_miRNA=[]
nonlog_no_target=[]
for i in range(0,len(miRNA)):
    col=miRNA.iloc[i]
    a_col=[]
    TE=2**col.logfold_TE
    a_col.append(col.RefseqAccession)
    a_col.append(TE)
    nonlog_miRNA.append(a_col)
    
    

for i in range(0,len(no_target)):
    col=no_target.iloc[i]
    a_col=[]
    TE=2**col.logfold_TE
    a_col.append(col.RefseqAccession)
    a_col.append(TE)
    nonlog_no_target.append(a_col)
nonlog_miRNA=pd.DataFrame(nonlog_miRNA,columns=["RefseqAccession","TE"])
nonlog_no_target=pd.DataFrame(nonlog_no_target,columns=["RefseqAccession","TE"])
nonlog_no_target
```

```python
no_target["mi_target"]="non"
no_target
```

```python
miRNA["mi_target"]="target"
miRNA
```

```python
join=miRNA.append(no_target)
join.to_csv("mir155_logfold.csv",index=False)
```
