```python
import pandas as pd
```

```python
#YTHDF の翻訳効率 miRNA1、mIRNA155の翻訳効率のデータ取得
m6A=pd.read_csv("/Users/miyakesho/Google_Drive/master/research/downloads/GSE63591_C-Y1-Ribosome_profiling.csv")
miRNA_1=pd.read_csv("/Users/miyakesho/Google_Drive/master/research/csv/symboled_miRNA1.csv")
miRNA_155=pd.read_csv("/Users/miyakesho/Google_Drive/master/research/csv/symboled_miRNA155.csv")
display(m6A.head(10))
display(miRNA_1.head(10))
```

```python
#m6AとmiRNAのデータをそれぞれ結合
m6A_miRNA1_merge_list=pd.merge(m6A, miRNA_1, on='Gene_symbol')
m6A_miRNA155_merge_list=pd.merge(m6A, miRNA_155, on='Gene_symbol')
print(len(m6A_miRNA1_merge_list),len(m6A_miRNA155_merge_list))
```

```python
display(m6A_miRNA1_merge_list.head(10))
display(m6A_miRNA155_merge_list.head(10))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

miRNAのターゲットとなっているgeneに関して色分け

```python
miRNA1_target=[]
miRNA155_target=[]
miRNA1_target_df=pd.read_csv("/Users/miyakesho/Google_Drive/master/research/csv/miRNA1_target_list.csv")
miRNA155_target_df=pd.read_csv("/Users/miyakesho/Google_Drive/master/research/csv/miRNA155_target_list.csv")
for i in miRNA1_target_df.Ref_seq:
    miRNA1_target.append(i)
for i in miRNA155_target_df.Ref_seq:
    miRNA155_target.append(i)
miRNA1_target[0:10]
```

```python
YTHDF_target=[]
YTHDF_target_df=pd.read_csv("/Users/miyakesho/Google_Drive/master/research/csv/PARCLIP_target.csv")
for i in YTHDF_target_df["Gene symbol"]:
    YTHDF_target.append(i)
YTHDF_target[0:10]
```

```python
def plot(x,y,col,name):
    fig, ax = plt.subplots(figsize=(10, 10))
    ax.scatter(x, y,color=col)
    ax.set_xlim(-4,4)
    ax.set_ylim(-4, 4)
    ax.vlines(0, -4, 4)
    ax.hlines(0, -4, 4)
    ax.set_xlabel('miRNA Translation Efficiency')
    ax.set_ylabel('m6A Translation Efficiency')
    plt.savefig("{}_TE.png".format(name),dpi=300)
```

```python
#散布図,ヒストグラムの作成
def make_fig(x_1,y_1,x_2,y_2,color,name,status):
    binnum=25
    fig, ax = plt.subplots(figsize=(10, 10))
    ax.scatter(x_1,y_1, c=color, label=name)
    ax.scatter(x_2,y_2, c='orange', label='non target')
    ax.set_xlim(-4,4)
    ax.set_ylim(-4, 4)
    ax.vlines(0, -4, 4)
    ax.hlines(0, -4, 4)
    ax.tick_params(labelsize=15)
    ax.legend(loc='upper right',fontsize=20)
    ax.set_xlabel('log(siYTHDF1/siControl)',fontsize=20)
    ax.set_ylabel('log(miRNA over expression/Control)',fontsize=20)
    plt.savefig("{}_{}_logfold.png".format(status,name),dpi=300)
    
    fig, ax = plt.subplots(figsize=(10, 4))
    ax.hist(x_1, color=color, label=name,density=True,alpha=0.5,bins=binnum)
    ax.hist(x_2, color='orange', label='no target',density=True,alpha=0.5,bins=binnum)
    ax.tick_params(labelsize=15)
    ax.set_xlim(-4,4)
    ax_yticklocs = ax.yaxis.get_ticklocs()#目盛りの情報を取得
    ax_yticklocs = list(map(lambda x: x * len(range(-3,3))*1.0/binnum, ax_yticklocs))#元の目盛りの値にbinの幅を掛ける
    ax.yaxis.set_ticklabels(list(map(lambda x: "%0.2f" % x, ax_yticklocs)))#直した目盛りを表示
    ax.set_ylabel('frequency',fontsize=20)
    plt.savefig("{}_{}_m6A_hist.png".format(status,name),dpi=300)
    
    fig, ax = plt.subplots(figsize=(10, 4))
    ax.hist(y_1, color=color, label=name,density=True,alpha=0.5,bins=binnum)
    ax.hist(y_2, color='orange', label='no target',density=True,alpha=0.5,bins=binnum)
    ax.tick_params(labelsize=15)
    ax.set_xlim(4,-4)
    ax.set_ylabel('frequency',fontsize=20)
    ax_yticklocs = ax.yaxis.get_ticklocs()#目盛りの情報を取得
    ax_yticklocs = list(map(lambda x: x * len(range(-3,3))*1.0/binnum, ax_yticklocs))#元の目盛りの値にbinの幅を掛ける
    ax.yaxis.set_ticklabels(list(map(lambda x: "%0.2f" % x, ax_yticklocs)))#直した目盛りを表示
    plt.savefig("{}_{}_miRNA_hist.png".format(status,name),dpi=300)
```
