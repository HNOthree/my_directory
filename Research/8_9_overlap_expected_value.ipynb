```python
import pandas as pd
```

```python
rep1_ythdf1_list=pd.read_csv("/Users/miyakesho/Google_Drive/master/research/csv/PARCLIP_1.csv")
rep2_ythdf1_list=pd.read_csv("/Users/miyakesho/Google_Drive/master/research/csv/PARCLIP_2.csv")
ythdf1_list=rep1_ythdf1_list.append(rep2_ythdf1_list)
ythdf1_list.head(10)
```

```python
window=0
while(window<=500):
    seed=7
    count=0
    for i in range(len(ythdf1_list)):
        row=ythdf1_list.iloc[i]
        start=row.Start
        end=row.End
        count+=(end-start+2*window-seed+1)
    #print("window",window,"expected count",count/(4**seed))
    print("[",window,",",count/(4**seed),"]")
    window+=25
```
