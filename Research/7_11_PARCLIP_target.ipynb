二つのPARCLIPのデータからターゲットとなるgene symbolを選出する

```python
import pandas as pd
```

```python
rep1_pd=pd.read_csv("/Users/miyakesho/Google_Drive/master/research/csv/GSE63591_PARCLIP_rep1.csv")
rep2_pd=pd.read_csv("/Users/miyakesho/Google_Drive/master/research/csv/GSE63591_PARCLIP_rep2.csv")
rep1_pd.head(10)
```

```python
both=[]
rep1=[]
for gene in rep1_pd["Gene symbol"]:
    rep1.append(gene)

for gene in rep2_pd["Gene symbol"]:
        if gene in rep1:
            both.append(gene)
both
```

```python
print(len(both),len(rep1_pd),len(rep2_pd))
```

```python
both=pd.Data
both.to_csv("PARCLIP_target.csv",index=False)
```
