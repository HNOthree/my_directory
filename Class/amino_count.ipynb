```python
def dictionary_make(length):
    dictionary={}
    amino=["A","C","D","E","F","G","H","I","K","L","M","N","P","Q","R","S","T","V","W","Y"]
    peptide=["A","C","D","E","F","G","H","I","K","L","M","N","P","Q","R","S","T","V","W","Y"]
    while (len(peptide[0])<length):
        new_peptide=[]
        for seq in peptide:
            for base in amino:
                new_peptide.append(seq+base)
        peptide=new_peptide
    for sequence in peptide: 
        dictionary[sequence]=0
    return dictionary
```

```python
def dictionary_count(proteins):
    for i in range(0,len(proteins)-length):
        peptide=proteins[i]
        for j in range(1,length):
            peptide+=proteins[i+j]
        protein_dictionary[peptide]+=1
            
```

```python
length=2
protein_dictionary=dictionary_make(length)
```

```python
k=1
all_length=0
for record_id, record in records.items():
    for feature in record.features:
        if feature.type=="CDS":
            try:
                protein_sequence=feature.qualifiers["translation"]
                protein_sequence=protein_sequence[0]
                all_length+=len(protein_sequence)
                dictionary_count(protein_sequence)
                k+=1
            except  KeyError:
                continue
               
```
