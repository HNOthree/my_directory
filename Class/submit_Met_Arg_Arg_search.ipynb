```python
path_to_genbank_file = ("/Users/miyakesho/Google_Drive/master/Class/生物情報実験法/python/task1/GCF_000231385.2_ASM23138v3_genomic.gbff") # 自分のファイルに書き換えてください
records = SeqIO.index(path_to_genbank_file, 'gb')
```

```python
record = records[list(records.keys())[0]]
bacteria_sequence=record.seq
```

```python
def MMA_search(protein):
    location=[]
    for i in range(len(protein)-2):
        motif=protein[i]+protein[i+1]+protein[i+2]
        if(motif=="MRR"):
            location.append(i)
        else:
            continue
    return location
```

```python
def convert(sequence):
    
    i=len(sequence)-1
    convert_seq=""
    while(i>=0):
        letter=sequence[i]
        if(letter=="A"):
            convert_seq+="T"
        elif(letter=="T"):
            convert_seq+="A"
        elif(letter=="G"):
            convert_seq+="C"
        elif(letter=="C"):
            convert_seq+="G"
        else:
            print("ERROR")
        i-=1
    
   
    return convert_seq
```

```python

all_length=0
LIST=[]
for record_id, record in records.items():
    for feature in record.features:
        if feature.type=="CDS":
            try:
                protein_sequence=feature.qualifiers["translation"]
                protein_sequence=protein_sequence[0]
                all_length+=len(protein_sequence)
                MMA_location=MMA_search(protein_sequence)
                if(len(MMA_location)>0):
                    LIST.append([feature.location.start,feature.location.end,MMA_location,feature.strand])
                #if(1==100):
                    #break
            except  KeyError:
                continue
               
```

```python
start_codon=["ATG","GTG","ATA","TTG"]
count_dict={}
for i in LIST:
    #print(i)
    sequence=bacteria_sequence[i[0]:i[1]]
   
    for j in i[2]:
        
        if(i[3]==-1):
            print(len(sequence)/3,j)
            sequence=convert(sequence)
            print(len(sequence)/3,j)
        startposition=int(j)*3
        startcodon=sequence[0:3]
        if(startcodon  in start_codon):
            print(sequence[0:3],j,len(sequence)/3)
            MRR=str(sequence[startposition:startposition+9])
        #print(MRR[6:])
            if MRR[6:] not in count_dict:
                count_dict[MRR[6:]]=1
            else:
                count_dict[MRR[6:]]+=1
```
