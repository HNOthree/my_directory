```python
path_to_genbank_file = ("/Users/miyakesho/Google_Drive/master/Class/生物情報実験法/python/task1/GCF_000231385.2_ASM23138v3_genomic.gbff") # 自分のファイルに書き換えてください
records = SeqIO.index(path_to_genbank_file, 'gb')
```

```python
def dictionary_make(length):
    dictionary={}
    DNA=["A","T","G","C"]
    kmer=["A","T","G","C"]
    while (len(kmer[0])<length):
        new_kmer=[]
        for seq in kmer:
            for base in DNA:
                new_kmer.append(seq+base)
        kmer=new_kmer
    for sequence in kmer: 
        dictionary[sequence]=0
    return dictionary
```

```python
def get_seq():
    path_to_genbank_file = ("GCF_000231385.2_ASM23138v3_genomic.gbff") 
    records = SeqIO.index(path_to_genbank_file, 'gb')
    record = records[list(records.keys())[0]]
    bacteria_sequence=record.seq
    
    return bacteria_sequence
```

```python
def kmer_count(kmer_dict,length,DNA):
    for i in range(0,len(DNA)-length):
        seq=DNA[i:i+length]
        kmer_dict[seq]+=1
    return kmer_dict
```

```python
def dictionary(length):

    genome_sequence=get_seq()
    genome_dictionary=dictionary_make(length)
    genome_dictionary=kmer_count(genome_dictionary,length,genome_sequence)
    CDS_dictionary=dictionary_make(length)
    tRNA_dictionary=dictionary_make(length)
    CDS_sumation=0
    tRNA_sumation=0
    
    for record_id, record in records.items():
        for feature in record.features:
            if feature.type=="CDS":
                CDS_sequence=genome_sequence[feature.location.start:feature.location.end]
                CDS_sumation+=((feature.location.end-feature.location.start)-length+1)
                CDS_dictionary=kmer_count(CDS_dictionary,length,CDS_sequence)
            elif feature.type=="tRNA":
                tRNA_sequence=genome_sequence[feature.location.start:feature.location.end]
                tRNA_sumation+=((feature.location.end-feature.location.start)-length+1)
                tRNA_CDS_dictionary=kmer_count(tRNA_dictionary,length,tRNA_sequence)
            
    return genome_dictionary,CDS_dictionary,tRNA_dictionary,CDS_sumation,tRNA_sumation
    
```

```python
genome_dictionary,CDS_dictionary,tRNA_dictionary,CDS_sum_length,tRNA_sum_length=dictionary(6)
```

```python
kmer_df=[]
for i in genome_dictionary:
    genome_count=genome_dictionary[i]
    CDS_count=CDS_dictionary[i]
    if(genome_count==0):
        kmer_df.append([i,0])
    elif(CDS_count==0):
        kmer_df.append([i,-sys.max])
    else:
        
        ratio=math.log2(CDS_count/genome_count)
        kmer_df.append([i,ratio])
kmer_df=pd.DataFrame(kmer_df,columns=["kmer","log2fold"])
```
